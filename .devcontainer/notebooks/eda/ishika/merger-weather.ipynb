```python
years = range(2014, 2024)
months = [str(i).zfill(2) for i in range(1, 13)]
```

```python
cols = ['Unnamed: 0', 'temp_avg', 'dew_avg', 'hum_avg', 'wind_speed_avg', 'pressure_avg', 'precip', 'month', 'year', 'location']
```

```python
# samut prakan
for year in years:
    for month in months:
            filename = f"datasets/Bangkok-university/a_{year}-{month}_weather.csv"
            try:
                # for every month of the year, i want to merge the data into one file
                # so 2014 should have one file with 12 months of data, etc...
                df = pd.read_csv(filename)

                # add column names
                df.columns = cols
                
                # add day column that starts like 01, 02, 03, etc...
                # day should start from 01
                df['day'] = df.index + 1

                df.to_csv(f"datasets/Bangkok-university/a_{year}-weather.csv", mode='a', header=None)
            except FileNotFoundError:
                print(f"File {filename} not found")
                continue
```

```python
cols = ['Unnamed: 0', 'Unnamed: 1', 'temp_avg', 'dew_avg', 'hum_avg', 'wind_speed_avg', 'pressure_avg', 'precip', 'month', 'year', 'location','day']
```

```python
def merge_files(year):
    df = pd.read_csv(f"datasets/Bangkok-university/a_{year}-weather.csv", header=None)
    df.columns = cols
    df.drop(columns=["Unnamed: 0", "Unnamed: 1"], inplace=True)
    return df
```

```python
prakan_2014 = merge_files(2014)
prakan_2015 = merge_files(2015)
prakan_2016 = merge_files(2016)
prakan_2017 = merge_files(2017)
prakan_2018 = merge_files(2018)
prakan_2019 = merge_files(2019)
prakan_2020 = merge_files(2020)
prakan_2021 = merge_files(2021)
prakan_2022 = merge_files(2022)
prakan_2023 = merge_files(2023)
```
