```python
locations = ['samut prakan', 'chulalongkorn-hospital', 'samut sakhon', 'thonburi-power',]
prefix = './sub_stations/'
x = []
for location in locations:
    globals()[location] = pd.read_csv(prefix+location+'.csv')
    globals()[location]['location'] = location 
    x.append(globals()[location])
x=pd.concat(x)
x.rename(columns={' pm25':'pm25', ' pm10':'pm10', ' o3':'o3',' no2':'no2', ' so2':'so2', ' co':'co'}, inplace=True)
cols = ['pm25','pm10', 'o3', 'no2','so2','co']
x[cols]=x[cols].apply(lambda x: x.replace(' ', np.NAN), axis=1)
x['pm25'] = x['pm25'].astype('float')
x['pm10'] = x['pm10'].astype('float')
x['co']=x['co'].astype('float')
x['date']=pd.to_datetime(x['date'])
x['year']=x['date'].dt.year
x['month'] = x['date'].dt.month
x['day'] = x['date'].dt.day
```

```python
strip = "%Y-%m-%d"
def iterative_merge(volume,start, end, location, idx, year, month, day):
    start = datetime.strftime(start , strip)
    x_ = x[x['location'] == location]
    c_year = x_[(x_['year'] == year) & (x_['month'] == month) & (x_['day'] == day)]
    if c_year.shape[0] > 0:
        t_x.loc[idx,'pm25'] = x[x['location'] == location].loc[str(start), 'pm25']
t_x['index']=t_x.index
# x.set_index('date', inplace =True).sort_index()
t_x.apply(lambda x: iterative_merge(x['volume'],x['start'], x['end'], 
                                    x['location'], x['index'], x['year'], x['month'], x['day']), axis=1)
```

```python
#taking distance for at least 8 km which approx 5 miles
t=t_x[t_x['distance'] < 4]
locations = t_x['location'].unique().tolist()
```

```python
stations_lat_long = {'latitude':[], 'longitude':[], 'location':[]}
gis = GIS()
def get_lat_long_by_state(location):
    if location  != None or location != '' :
        geolocator = Nominatim(user_agent="myGeopyApp/1.0")
        esrihq = geocode(location)
        if esrihq != None:
            latitude, longitude = esrihq[0]['location']['x'], esrihq[0]['location']['y']
            return latitude, longitude
for loc in locations:
    longitude, latitude = get_lat_long_by_state(loc)
    stations_lat_long['latitude'].append(latitude)
    stations_lat_long['longitude'].append(longitude)
    stations_lat_long['location'].append(loc)
```

```python
# This makes sense if you believe that the relationship between traffic and pollution is 
# spatially or temporally structured. By shuffling traffic values across nearby locations, 
# you are testing whether the observed correlation is a result of spatial proximity between 
# traffic and pollution levels, or if it's simply due to random noise.


for location in locations[:-1]:
    t_ = t[t['location'] == location]
    observed_corr, p_value = spacial_temporal_permutation_test.spatial_permutation_test(t_['pm25'].values, 
                                            t_['event_count'].values, 
                                            t_[['latitude', 'longitude']].values)
    print(f"Observed Correlation for {location}: {observed_corr:.4f}")
    print(f"P-value: {location} {p_value:.4f}")
```

Samut Prakan:

    Located just south of Bangkok, it is part of the Bangkok Metropolitan Area.
    The province has high population density and is heavily influenced by the urban sprawl of Bangkok.
    Samut Prakan is home to several industrial zones and ports, including Laem Chabang Port, one of the busiest seaports in Thailand. This contributes to significant air pollution from transportation and industrial emissions (factories, shipping, and truck traffic).
    The road traffic in Samut Prakan is significant due to its proximity to Bangkok, where high traffic congestion leads to increased PM2.5 and nitrogen dioxide (NO2) emissions.

Samut Sakhon:

    Located southwest of Bangkok, it is known for its agriculture (especially shrimp farming) and fishing industry.
    Samut Sakhon also has a large industrial base, particularly in the food processing industry, but it is less industrialized than Samut Prakan.
    There is some road traffic and transportation impact, but the overall urban sprawl and congestion are lower compared to Samut Prakan.
    The pollution sources in Samut Sakhon are somewhat less diversified than in Samut Prakan, but industrial pollution (especially from food processing) and vehicle emissions can still contribute to air quality issues.
