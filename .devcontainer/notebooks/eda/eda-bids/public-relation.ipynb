```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../../../../data/raw/dataset-bids/public-relations department, bangkok-air-quality.csv', parse_dates=['date'])
```

```python
df.head()
```

```python
# trim white space of column names
df.columns = df.columns.str.strip()
```

```python
def fill_numeric(col):
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
fill_numeric('pm25')
fill_numeric('pm10')
fill_numeric('o3')
fill_numeric('no2')
fill_numeric('so2')
fill_numeric('co')
```

```python
# heatmap of missing value for pm25
plt.figure(figsize=(12, 8))
sns.heatmap(df.isna(), cbar=False)
plt.show()
```

```python
# correlation matrix
sns.heatmap(df.drop(columns=['date']).corr(), annot=True)
```
